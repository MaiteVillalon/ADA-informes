###***1. PROBLEMA DE ORDENAMIENTO***###

**Entrada**: Arreglo de n números o lista de n números [n1, n2, n3 ...n].

**Salida**: El arreglo o la lista de números de entrada ordenada de forma ascendente, obteniendo [n1<n2<n3...n].


Cuando se desarrollaban los primeros programas, en algun momento surgió la necesidad de manejar los datos de forma ordenada, ya sea en orden ascendente o descendente. Desde ese momento comenzo la busqueda de algun tipo de algortimo que diera solucion a este problema. En la actualidad contamos con más de un par de algoritmos que nos ayudan de forma eficiente a realizar el ordenamiento, en este informe analizaremos uno en especifico, **Insertion Sort**.

###***2.  DESCRIPCIÓN DEL ALGORITMO***###




El algoritmo **Insertion Sort** recibe como parametro una lista o arreglo de N datos que el programador desea ordenar. No se necesita una lista o arreglo auxiliar, los números se van ordenando en la misma lista o arreglo que se ingresó.

***Pasos importantes:***


1.   El algoritmo irá analizando uno a uno los números del arreglo, por cada uno de ellos se preguntará si es menor al numero que se encuentra en la posicion anterior, si es menor se intercambiaran de posición, se irá comparando con cada número que este antes en el arreglo, de esta forma por cada iteracion (analizar **cada numero)**, se obtendrá un **subarreglo**
 ordenado.  
 
2.   Si el arreglo esta ordenado, el algoritmo no volvera a iterar.

3.   Al finalizar, el algoritmo retornará el arreglo de entrada ordenado de forma ascendente.



***Ejemplo:***

El algortimo recibirá el siguiente arreglo de números enteros: [ 7 - 2 - 6 - 14 ]



*   En la primera iteracion ( i = 1 ) el algoritmo analizara el elemento del arreglo en tal posicion i , se consultara si es menor a el/los elemento(s) antes de el, tenemos que 2 < 7, por ende intercambiaran posiciones obteniendo el siguiente subarreglo [ **2** - **7** - 6 - 14 ]. 

*   En la siguiente iteracion se analizara el número 6 ( i = 2 ), y se comparará con los números en las posiciones anteriores a este, primero se compará con el número 7, tenemos que (6 < 7), por ende intercambian posiciones, posteriormente se comparará con el numero 2, tenemos que ( 6 > 2 ), por ende **no** intercambian posiciones. Como ya llegamos a la posición 0 del arreglo, se deja de comparar y obtenemos el siguiente arreglo 
[ **2** - **6** - **7** - 14 ].    
*   En la siguiente iteración se analizara el último número del arreglo ( 14 ), se comienza comparando con el número en la posicion anterior a este ( 7 ), tenemos que ( 14 > 7 ), por ende no intercambiaran posiciones, debido a que en cada iteracion el arreglo va quedando ordenado desde la posicion i (numero que se analiza) hasta el numero en la posición i = 0, el numero 14 no será mayor a ningun número que se encuentre en una posición anterior a donde se encuentra el numero 7.


*   El algoritmo termina y retorna el arreglo ordenado **[ 2 - 6 - 7 - 14 ]** y la cantidad de intercambios. 









###***3. IMPLEMENTACION DEL ALGORITMO***###

```python
import random
from termcolor import colored
import copy


def insertion_sort(arr, verbose=False):
    n = len(arr)
    cont = 0
    swap = 0
    for i in range (1, len(arr)):
      aux = arr[i]
      k = i-1
      cont = cont + 1

        

      while k >= 0 and aux < arr[k]:
            arr[k+1] = arr[k]
            k = k-1
            swap = swap +1


            

      arr[k+1] = aux
      if verbose == True:
          print("Pasada" , i )
          print(arr)
          print("cantidad de intercambios realizados en total : ",swap)
          print("\n")
      
    return arr ,swap


  
arr = [7,6,14,2,5]
insertion_sort(arr)
print("\nArreglo ordenado de forma ascendente: ")
print(arr)


```

##***3.1 Ejecución del algoritmo paso a paso (verbose = True)***##
 

```python
arr =  [5,2,4,6,1,3]
insertion_sort(arr, verbose = True)  
```

Podemos notar que en cada pasada (iteración), se va ordenando el arreglo, tomando cada uno de los numeros de el array de entrada.

Podemos distinguir que en cada pasada el arreglo se va ordenando desde la posición i = 0 , hasta la posición del numero a ordenar.


###***4. TIEMPO DE EJECUCIÓN***###

####**Teorema ( Tiempo de ejecución )**###

El **tiempo de ejecución** del algoritmo **Insertion Sort** es de **O(n²**) en el peor caso. Lo cual significa que el tiempo que tomaría ordenar el array seria equivalente al **cuadrado de la cantidad de elementos** del array.


##Prueba del Teorema ( Tiempo de ejecución )###

El peor caso de Insertion Sort será cuando el array de entrada este ordenado de mayor a menor, donde por cada elemento n debo realizar n-1 intercambios. A continuación observaremos el algortimo entregandole un array en orden descendente.


**Array de entrada = [7, 6, 5, 4]**



```python
arr = [7, 6, 5, 4,]
insertion_sort(arr, verbose = True)
```


En el algortimo anterior podemos ver como se va actualizando la cantidad de intercambios realizado segun cuantos se hayan echos en cada pasada.

Podemos observar que en la **pasada 1**  se reviso el numero arr[1]  y se realizo **1 intercambio**. 

En la **pasada 2** se reviso el numero arr[2] y se realizaron **2 intercambios**.

En la **pasada 3** se reviso numero arr[3] numero y se realizaron **3 intercambios**, obteniendo un total de **6 intercambios**.

El caso promedio debe cumplir con [n*(n-1)]/2 comparaciones, por ende el tiempo de ejecución del algoritmo en el peor caso es O(n²).




##***5. CORRECTITUD***###

###**Teorema (Correctitud)**

El algortimo **Insertion Sort** entregara un arreglo ordenado de forma ascende con los mismos datos del arreglo entregado por el usuario.

###Prueba del teorema
Analizando el codigo anterior, podemos notar que por cada pasada i que realizó, los elementos del arreglo desde **i - 1** hasta **i = 0** estaran ordenados en orden ascendente.

A continuacion demostraremos la siguiente **propiedad invariante de bucle** del algoritmo Insertion Sort:



*   "Al comiezo de cada pasada  i (iteración), obtendremos un subarreglo ordenado desde la posición ( i -1 ) hasta ( i = 0 ), con los mismos elementos en esas posiones del arreglo de entrada".

**Comienzo**

En la primera pasada (i = 1), tenemos que **desde ( i - 1 ) hasta (1 = 0)** veremos un subarreglo ordenado [7] con los mismo numeros en esas posiciones del arreglo de entrada. Todos los elementos anteriores a i cumplen con la propiedad.

**Mantención**

Podemos concluir que para todas las interaciones, tendremos que al comienzo se cumplira la propiedad desde ( i - 1 ) hasta  ( i = 0 ).
