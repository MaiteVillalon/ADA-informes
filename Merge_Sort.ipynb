###***Merge Sort***

El algoritmo de ordenamiento **Merge Sort** funciona bajo la frase "*divide y venceras*". Recibe por entrada un arreglo de n numeros y entrega tal arreglo ordenado de forma ascendente. Este algoritmo se implementa de forma recursiva. 

**Pasos Importantes**


1.   El algoritmo divide el arreglo en dos subarreglos.
2.   Cada subarreglo se vuelve a dividir a la mitad en otros subarreglos, y asi se itera hasta que obtengamos solo arreglos que contengan un solo elemento.
1.   El algoritmo ordena cada subarreglo de manera recursiva. 
2.   La funcion merge combina los dos subarreglos ya ordenados en un solo arreglo.



### ***Función Merge*** 

La función merge es parte fundamental del algoritmo, es la encargada de unir en un solo arreglo las dos subsecuencias ordenadas.

La función recibe 4 parámetros: 



*   A: el arreglo.
*    izq,der,verbose: Los dos subarreglos y opción verbose.



Se confia en que los sub-arreglos izq[m..l] y der[l+1..t] se encuentran previamente ordenados de menor a mayoe. Posteriormente la función los combina generando un sólo arreglo ordenado A[m..t].


   **Pasos Importantes**


1.   Primero, se crea un nuevo arreglo para guardar los elementos ordenados (B).
2.   Se compara el primer elemento de cada subarreglo del arreglo A, se escoge el menor entre los dos y se coloca en la primera posicion del arreglo B.
1.  El elemento escogido en el paso anterior es eliminado de su subaareglo correspondiente.
2.  Se repetirá el paso 2 hasta que uno de los dos subarreglos de A quede sin elementos. En ese momento se copian todos los elementos que quedan en el subarreglo **no vacio** en el arreglo B.

1.   Finalmente se copia el contenido del arreglo B, en el arreglo principal A.

##**3. Implementación del Algoritmo**

```python
def mergeSort(arr,contador, verbose):
    
    # Pregunto si el arreglo tiene más de un elemento
    if len(arr) > 1:

        # Calculo cual es la mitad del array
        #mitad = len(arr)//2
        mitad = len(arr)//2

        izq= (arr[:mitad])
        der=(arr[mitad:])
 
        # Con estas llamadas a MergeSort se ordenan los dos subarreglos 
        #  que se subdividio arr y luego se llama a la funcion merge para
        #  que los combine de forma ordenada.
        contador[0]= contador[0]+1
        mergeSort(izq, contador,verbose )
        mergeSort(der,contador ,verbose )
        arr[:]=merge(izq,der,contador, verbose)
 
 
# Driver code to test above
arr = [5,8,7,1,3]
n = len(arr)
print("Arreglo de entrada")
for i in range(n):
    print("%d" % arr[i],end=" ")
print("\n") 
contador = [0,0]
mergeSort(arr,contador, verbose= False)
print("\n\nArreglo ordenado")
for i in range(n):
    print("%d" % arr[i],end=" ")

print("\n")
print("cantidad de recursiones:" , contador[0])
print("cantidad de comparaciones:" , contador[1])
```

```python

def merge(izq, der, contador, verbose):
    
    i, j = 0, 0 # Variables de incremento
    result = [] # Lista de resultado
 
   # Se elige el menor de los subarreglos
    while(i < len(izq) and j < len(der)):
        if (izq[i] < der[j]):
            result.append(izq[i])
            i += 1
            contador[1]= contador[1]+1
        else:
            result.append(der[j])
            j += 1
            contador[1]= contador[1]+1
            
 
   # Agregamos los resultados a la lista
    result += izq[i:]
    result += der[j:]
    contador[1]= contador[1]+1
 
    # Retornamos el resultados
    return result
 
```

```python
def merge(izq, der,contador, verbose ):
    
    i, j = 0, 0 # Variables de incremento
    result = [] # Lista de resultado
 
   # Intercalar ordenadamente
    while(i < len(izq) and j < len(der)):
        if (izq[i] < der[j]):
          if verbose == True:
            print("Se comparan los siguientes elementos:\n")
            print(der[j], "es más mayor que", izq[i])
            print("\n")
          result.append(izq[i])
          i += 1
          contador[1]= contador[1]+1
        else:
          if verbose == True:
            print("Se comparan los siguientes elementos:\n")
            print(der[j], "es más mayor que", izq[i])
            print("\n")
          result.append(der[j])
          j += 1
          contador[1]= contador[1]+1
 
   # Agregamos los resultados a la lista
    result += izq[i:]
    result += der[j:]
    contador[1]= contador[1]+1
 
    # Retornamos el resultados
    if verbose == True:
      print(result)
    return result
 
```

***Teorema (Tiempo de ejecución)***

El algoritmo MergeSort presenta un tiempo de ejecución igual a O(nlogn), en todos sus casos.

Cuando tenemos más de un elemento en nuestro array consideramos los siguientes tiempos de ejecución para los pasos que se presentana continuación.

**Dividir:** Cuando el algortimo divide el arreglo en dos subarreglos toma un tiempo constante de O(1).

**Recursividad:**: De manera recursiva se ordenaran los dos subarreglos, el cual tomara un tiempo de 2T(n/2).

**Unir:** Realizar el procedimiento Merge a un arreglo de n elementos tomara un tiempo de O(n).

##**Correctitud**

***Teorema(Función Merge)***

La función merge une dos subarreglos ya ordenados en un solo arreglo ordenado de manera ascendete



***Propiedad invariante de bucle (función Merge)***

Al comienzo de cada iteracion de nuestro ciclo while de la funcion merge, nuestro arreglo result tendrá los elementos más pequeños de los subarreglos izq y der ordenados de forma ascendente.

Debemos demostrar que esta propiedad de bucle es correcta durante todo el algoritmo.

**Inicialización :** Antes de la primera iteracion, tenemos que el arreglo result, esta vacío, por lo cual consideramos que se encuentra ordenado.

**Mantención:** En las siguientes iteración se le añadira izq[i] o der[j], dependiendo cual de ellos sea menor, sabemos que el ultimo elemento añadido a result es menor a izq[i] y der[j], por ende los elementos que se vayan a insertar seran siempre mayores.

**Finalización:** Una vez se termine ed recorrer uno de los subarreglos, se copiara el contenido del otro en el arreglo result.




***Correctitud MergeSort***

Para probar la correctitud del algortimo MergeSort usaremos inducción matemática.

Para que nuestra declaración sea correcta debemos probar que sea correcta en dos casos:


1.   Caso base
2.   Casos inductivos


**Hipotesis:** El algoritmo MergeSort entrega el mismo arreglo de entrada de n elementos, ordenado de menor a mayor.

Nuestro caso base sera cuando n = 1, por ende el algoritmo retornará el mismo elemento, por lo cual se cumple la hipotesis para este caso.

Para el caso inductivo, tenemos que el algortimo va a dividir el arreglo de entrada en dos subarreglos (a,b), cada uno de **tamaño n/2** y posteriormente ordenara cada uno de ellos llamando recursivamente a mergesort, **cada uno de sus tamaños es menor a n**, por ende asumimos que los arreglos se ordenarán de forma correcta, y finalmente la función Merge los unira en un solo arreglo, la correctitud de la función merge fue probada anteriormente, por ende podemos decir que el algortimo MergeSort es correcto.
