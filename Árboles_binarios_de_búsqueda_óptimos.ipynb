###**Descripción del Problema**

El problema que trataremos en este informe es el de **Árboles binarios de búsqueda óptimos**, un problema de optimización, encontraremos la forma más optima de ubicar los nodos. 

Para resolver el problema se utiliza **programación dinámica**. Esta tecnica se utiliza cuando los subproblemas que resolvemos nos vuelven a aparecer durante el transcurso de la resolución, por ende, para no resolverlos cada vez que aparescan guardaremos la solución, evitando de esta forma realizar trabajo extra.

El fin de construir el **Árbol binario** es minimizar la cantidad de nodos visitados en cada busqueda.



```
Entrada: Secuencia ordenada de n claves: K=⟨k1,k2,...,kn⟩. Cada clave ki tiene una probabilidad pi 
de ser buscada. Existen claves ficticias di que representan búsquedas fallidas con ki−1≤di≤ki. Cada clave 
di tiene una probabilidad qi de ser buscada. La suma de probabilidades debe 
ser uno, es decir: ∑i=1npi+∑j=0nqi=1
```

```
Salida: Construir un árbol de búsqueda óptimo, es decir, un árbol que minimice la cantidad esperada de 
nodos visitados (costo esperado).
```



##**Descripción del Algoritmo**

Como mencionamos anteriormente, el problema del árbol de búsqueda óptimo, consiste en encontrar la forma más optima de ubicar los nodos de un árbol con el fin de **minimizar la cantidad promedio de nodos visitados por cada búsqueda**.

La cantidad promedio de nodos visitados por cada búsqueda se denomina **costo esperado**, y se calcula multiplicando el costo de cada nodo por su probabilidad de búsqueda.

Existen diferentes formas de implementar este algoritmo, en este informe veremos 2 de ellas, **recursiva y bottom up**.

###**BST RECURSIVO**

El algoritmo recibe por entrada un arreglo con las claves de los nodos **K = [ k1... , kn ]** , un arreglo con las probilidades de ocurrencia para cada clave **Q = [ q1... , qn ]** , un arreglo con las probabilidades para las claves ficticias **D = [ d1... , dn ]** y dos indices  **i**, **j** que en cada llamada recursiva indicaran la clave inicial y final del subarbol que se esta procesando.

*Pasos importantes a seguir:*


*   Si los indices del comienzo del arreglo es igual al final se retorna 0.
*   Calcular la suma de las probabilidades ( de nodos y claves ficticias ).

*   Cada clave del arreglo será elegida como raiz del árbol.
*   Recursivamente se encontrara el mejor subarbol izquierdo y derecho de la raiz escogida anteriormente.

*   Se calcula costo esperado.***
*   Se almacenan los costos minimos de cada árbol.


*   Se construye el arbol optimo con la raiz y con los subarboles con mínimo costo.
*Se retorna árbol final y costo.


Costo esperado: El costo esperado de cada árbol se calcula como la suma de el costo del arbol izquierdo, el costo del arbol derecho y sus probabilidades.

###**BST BOTTOM UP**


El algoritmo recibe por entrada un arreglo con las claves de los nodos **P = [ P1... , Pn ]** , un arreglo con las probilidades de ocurrencia para cada clave **Q = [ q1... , qn ]** y una variable n la cual indica el tamaño del arreglo.

*Pasos importantes a seguir:*


*   Se crean 3 matrices, e guarda el subarbol optimo, w guarda sus probabilidades y root va creando árbol binario más eficiente.
*  Se inicializan las matrices e y w con las probabilidades de las claves fallidas.
* La suma de las probabilidades se guarda en la matriz w
*Se calcula el costo esperado del árbol optimo con la raiz k.
*Se guarda el costo óptimo en la matriz e y se guarda la raíz que minimiza el costo en root.

````python
import numpy as np
from math import inf


def bst_recursivo(k,q,d,cont,com,fin,verbose):

  cont[0]+=1

  if com > fin:    #Caso base
    return d[com-1]


  #Suma probabilidades
  sum_prob_q = sum(q[com-1 : fin])
  sum_prob_d = sum(d[com-1 : fin + 1])
  prob_total = sum_prob_q + sum_prob_d




  #Se inicializa costo
  costo = inf

  for i in range (com, fin + 1):

    min_cost_izq = bst_recursivo(k,q,d,cont,com, i - 1,verbose) #Llamada recursiva subarobol izquierdo
    min_cost_der = bst_recursivo(k,q,d,cont,i + 1, fin,verbose) #Lamada recursiva subarbol derecho
    min = min_cost_izq + min_cost_der + prob_total #Calculo de costo mínimo

    if verbose == True:
      print("Valor del costo minimo: ", min)

    if min < costo:  #Si min es más pequeño que costo, variable costo se actualiza
       costo = min 
       if verbose == True:
         print("Se actualiza la variable costo")
    else:
        if verbose == True:
         print("Se mantiene el valor de la variable costo")


  return costo  #Se retorna el costo mínimo


k = [1,2,3,4,5] #Claves
q = [0.15, 0.10, 0.05, 0.10, 0.20] #Probabilidad claves 
d = [0.05, 0.10, 0.05, 0.05, 0.05, 0.10] #Propabilidad claves ficticias
cont = [0]
verbose= False
costo_esperado = round(bst_recursivo(k,q,d,cont,1,len(q),verbose),2)
print ("Costo esperado: ",costo_esperado)
print("Cantidad de llamadas recursivas: ",cont[0])
````

````python
import numpy as np
import math

def optimalBST(p,q,n,cont,verbose):

    #Se crean las matrices
    e = np.zeros((n+1)**2).reshape(n+1,n+1)
    w = np.zeros((n+1)**2).reshape(n+1,n+1)
    raiz = np.zeros((n+1)**2).reshape(n+1,n+1)

    # Se guardan las probabilidades
    for i in range(n+1):
        e[i,i] = q[i]
        w[i,i] = q[i]
    for i in range(0,n):
        raiz[i,i] = i+1

    if verbose == True:
      print("Se crean e inicializan matrices e y w")
      print("La matriz e almacena los costos y la matriz w sus probabilidades\n")
      print(e)
      print("\n")
      print(w)



    for l in range(1,n+1):
        for i in range(0, n-l+1):
            j = i+l
            min_ = math.inf
            w[i,j] = w[i,j-1] + p[j] + q[j]   #Se suman las probabilidades
            cont[0] += 1
            if verbose == True:
              print("\nSuma de las probabilidades:", round(w[i,j],3),"\n" )
            for r in range(i,j):
                t = e[i, r-1+1] + e[r+1,j] +  w[i,j]
                if t < min_:
                    min_ = t                
                    e[i, j] = t
                    raiz[i, j-1] = r+1
            if verbose == True:
              print("Matriz e ( costos mínimos ) ")
              print(e,"\n")
              print("Matriz w ( probabilidades ) ")
              print(w)

    return (e,raiz)

def main():

    p = [0.5  ,  0.1   ,  0.05]
    q = [0.15  ,  0.1 ,   0.05 ,   0.05]
    n = len(p)-1
    cont = [0]
    verbose=False

    e,raiz= optimalBST(p,q,n,cont,verbose)
    print("------ e -------")
    print(e)
    print("----- root -----")
    print(raiz)

    print("Cantidad de comparaciones: ", cont[0])

if __name__ == '__main__':
    main()
````

##**Propiedades del algoritmo**


Para demostrar el que el algoritmo bottom-up es correcto, debemos calcular el costo mínimo, lo que nos llevara una búsqueda bastante exhaustiva de todas las posibilidades, para eso etiquetamos los nodos de **k_1,k_2...k_n** para crear el árbol binario y agregarle las claves ficticias, esta búsqueda de árbol binario con n nodos tendíamos que examinar un número exponencial de árboles, por lo que es normal que debamos resolver este problema con programación dinámica.

**Sub estructura óptima**

Para la sub estructura optima debemos considerar que cualquier sub árbol ki...kj debemos saber que se debe cumplir que **1<=i<=j<=n**, donde los sub arboles deben contener claves ficticias **di−1,....dj**

Ahora sabemos que el árbol binario de búsqueda óptimo T tiene sub árboles que contienen las claves ki...kj con sus respectivas claves ficticias di−1,....dj podemos decir que **kr(i<=r<=j)** es sub raíz de una sub árbol óptimo que contiene todas las claves

El subárbol izquierdo contendría todas la clave **k_r ki...kr−1** y **di−1,....dr−1** y el sub árbol derecho contiene **kr+1...kj**.

Al examinar todos los candidatos kr y aseguraríamos que encontramos árbol binario de búsqueda óptimo mientras **i<=r<=j**

**Definir una función recurrente**

para definir nuestra función recurrente debemos saber nuestro dominio de subproblema que contenga las claves **ki...kj** cuando **1<=i<=j<=n**. Ahora podemos definir e[i,j] que es el costo esperado 

Captura de pantalla (147).png

La manera de calcular la suma de las probabilidades para cada sub árbol seria:

$w[i,j]=∑jl=ipl+∑jl=i−1ql$

De esta manera tenemos que en el caso de que Kr es una raíz de un árbol binario de búsqueda óptimo ki...kj tenemos que:

$e[i,j]=pr+(e[i,r−1]+w(i,r−1))+(e[r+1,j]+w(r+1,j))$

ahora podemos notar que:

$w(i,j)=w(i,r−1)+pr+w(r+1,j)$

ahora que notamos como queda w(i,j) podemos reescribir e[i,j]

$e[i,j]=e[i,r−1]+e[r+1,j]+w(i,j)$

**Inicialización:**

Al principio se crean y se inicializa la diagonal de la matriz e[] que contiene la suma de los costos mínimos, esto siempre será correcto porque aún no tenemos sub problemas

**Mantención:**

##**Experimentos**

````python
import random
import numpy as np
from timeit import repeat
import matplotlib.pyplot as plt

def optimal_bst_instance_generator(n):
    keys = sorted(random.sample(range(1, 100), n))
    arr = np.random.random(n*2+1)
    arr /= arr.sum()
    
    p = list(arr[:n]) # Probabilidad de las claves
    q = arr[n:] # Probabilidad de las claves ficticias
    return keys, p, q.tolist()
````

````python
x=[]; y1=[]; y2=[];

for n in range(5,10):
  keys, p, q = optimal_bst_instance_generator(n)
  cont = [0]
  verbose = False
  t1= repeat(setup="from __main__ import optimalBST", stmt=f"optimalBST({p},{q},{len(p)-1},{cont},{verbose})", repeat=1, number=10)
  t2 = repeat(setup="from __main__ import bst_recursivo", stmt=f"bst_recursivo({keys},{p},{q},{cont},{1},{len(p)},{verbose})", repeat=1, number=10)

  x.append(n)
  y1.append(t1)
  y2.append(t2)

plt.plot(x,y1)
plt.plot(x,y2)
plt.legend(["Bottom-Up", "Recursivo"])

plt.xlabel('n')
plt.ylabel('Tiempo en ms')
plt.show()
````
